## Simple feature changes

### Dropping the LEGAL_NAME of the merchant and dates for now

```python
merchants_og_df = merchants_og_df.drop(['LEGAL_NAME','FOUNDATION_DATE','ACTIVE_FROM','TRADING_FROM'], axis=1)
```

```python
merchants_og_df.dtypes
```

### Converting to categorical Data

```python
merchants_og_df = pd.get_dummies(merchants_og_df, columns = ['BUSINESS_TYPE','OUTLET_TYPE'])
```

```python
merchants_og_df.head()
```

```python
merchants_og_df.dtypes
```

```python
transactions_train_og = pd.read_csv('./data/transactions_train.csv', usecols=['TX_ID','MERCHANT_ID','TX_FRAUD'])
transactions_train_og = transactions_train_og.set_index("TX_ID")
```

```python
transactions_train_merged  = pd.merge(transactions_train_og,merchants_og_df,on='MERCHANT_ID')
transactions_train_merged.count()
```

```python
transactions_train_merged.set_index('MERCHANT_ID')
transactions_train_merged.head()
```

## Looking at some distributions


```python
sns.countplot(x='MCC_CODE', data=transactions_train_merged,hue = 'TX_FRAUD')
```

```python
transactions_train_merged.columns
```

```python
plot_columns = ['MCC_CODE', 'TAX_EXCEMPT_INDICATOR',
       'PAYMENT_PERCENTAGE_FACE_TO_FACE', 'PAYMENT_PERCENTAGE_ECOM',
       'PAYMENT_PERCENTAGE_MOTO', 'DEPOSIT_REQUIRED_PERCENTAGE',
       'DEPOSIT_PERCENTAGE', 'DELIVERY_SAME_DAYS_PERCENTAGE',
       'DELIVERY_WEEK_ONE_PERCENTAGE', 'DELIVERY_WEEK_TWO_PERCENTAGE',
       'DELIVERY_OVER_TWO_WEEKS_PERCENTAGE', 'BUSINESS_TYPE_Corporations',
       'BUSINESS_TYPE_Limited Liability Company (LLC)',
       'BUSINESS_TYPE_S Corporations', 'BUSINESS_TYPE_Sole Proprietorships',
       'OUTLET_TYPE_Ecommerce', 'OUTLET_TYPE_Face to Face',
       'OUTLET_TYPE_Face to Face and Ecommerce']
for col in plot_columns:
	sns.countplot(x=col, data=transactions_train_merged,hue = 'TX_FRAUD')
	plt.show()	
```

```python
fraudalent_cases = transactions_train_merged[(transactions_train_merged['TX_FRAUD'] == 1)]
authentic_cases = transactions_train_merged[(transactions_train_merged['TX_FRAUD'] == 0)]
```

```python
for col in plot_columns:
	sns.countplot(x=col, data=fraudalent_cases,color = 'orange')
	plt.show()	
	sns.countplot(x=col, data=authentic_cases,color = 'blue')
	plt.show()	
```

```python
sns.lineplot( x = authentic_cases.ANNUAL_TURNOVER_CARD.value_counts().index, y = authentic_cases.ANNUAL_TURNOVER_CARD.value_counts())
plt.show()
sns.lineplot( x = fraudalent_cases.ANNUAL_TURNOVER_CARD.value_counts().index, y = fraudalent_cases.ANNUAL_TURNOVER_CARD.value_counts())
```
